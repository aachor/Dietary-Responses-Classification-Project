```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the models
models = ['GPT 3.5', 'Gemini']
total_correct = [23, 24]  # Total number of correct classifications by each model
correct_safe = [22, 17]  # Correct safe classifications
correct_unsafe = [1, 7]  # Correct unsafe classifications

# X locations for the groups
x = np.arange(len(models))  
width = 0.10  # Width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, total_correct, width, color='#CD853F', label='Total Correct')
rects2 = ax.bar(x, correct_safe, width, color='grey', label='Correct Safe')
rects3 = ax.bar(x + width, correct_unsafe, width, color='#A9A9A9', label='Correct Unsafe')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Classifications')
ax.set_title('Comparison of Classification Performance by Model')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Function to attach a label above each bar, showing its height
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()
plt.show()
```

## GPT 3.5 Classifiaction Report

```python
# Mapping text labels to numeric values if analysis requires
label_mapping = {'Unsafe': 0, 'Safe': 1}
dataset['Annotator'] = dataset['Annotator'].map(label_mapping)
dataset['GPT 3.5'] = dataset['GPT 3.5'].map(label_mapping)

# Checking the mapping
dataset.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generating classification report
report_gpt = classification_report(dataset['Annotator'], dataset['GPT 3.5'], target_names=['Unsafe', 'Safe'])
print("Classification Report for GPT 3.5:")
print(report_gpt)

# Generating confusion matrix
cm_gpt = confusion_matrix(dataset['Annotator'], dataset['GPT 3.5'])
```

```python
# Function to plot confusion matrix
def plot_confusion_matrix(cm, classifier_name, labels):
    fig, ax = plt.subplots(figsize=(3, 3))
    cax = ax.matshow(cm, cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix for {classifier_name}')
    
    # Setting tick positions
    ax.set_xticks(range(len(labels)))
    ax.set_yticks(range(len(labels)))
    ax.set_xticklabels(labels)
    ax.set_yticklabels(labels)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('True')

    # Annotating each cell with the numeric value
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], 'd'), ha='center', va='center', color='red' if cm[i,j] > cm.max()/2 else 'black')

    plt.show()

# Plotting the confusion matrix
plot_confusion_matrix(cm_gpt, 'GPT 3.5', ['Unsafe', 'Safe'])
```

## Gemini Classification Report

```python
# Define the mapping from text labels to numeric labels
label_mapping = {'Unsafe': 0, 'Safe': 1, 'Neutral': None}

# Strip leading and trailing whitespace from the 'Gemini' column
dataset['Gemini'] = dataset['Gemini'].str.strip()

# Apply the label mapping to the 'Gemini' column
dataset['Gemini'] = dataset['Gemini'].replace(label_mapping)

# Now filter out rows where 'Gemini' is None (which were 'Neutral')
filtered_dataset = dataset[dataset['Gemini'].notna()]

# Convert 'Gemini' to integer after filtering out None
filtered_dataset['Gemini'] = filtered_dataset['Gemini'].astype(int)

# Verify changes and the filtering process
print("Data types after conversion:")
print(filtered_dataset.dtypes)  # This will show Gemini as int

print("Filtered dataset:")
print(filtered_dataset.head())
```

```python
# Counting the number of rows in the filtered dataset
num_rows = filtered_dataset.shape[0]
print(f"Number of rows after filtering: {num_rows}")
```

```python
# Generating confusion matrix for Gemini
cm_gemini = confusion_matrix(filtered_dataset['Annotator'], filtered_dataset['Gemini'])

# Plotting the confusion matrix for Gemini
plot_confusion_matrix(cm_gemini, 'Gemini', ['Unsafe', 'Safe'])
```

# CLEARER CONFUSION MATRIX

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_confusion_matrix(cm, classifier_name, labels):
    # Setting up the plot
    fig, ax = plt.subplots(figsize=(3, 3))
    # Using a lighter colormap; adjustable alpha
    cax = ax.matshow(cm, cmap=plt.cm.Blues, alpha=0.3)  
    
    # Title and labels
    plt.title(f'Confusion Matrix for {classifier_name}')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('True')

    # Setting tick positions
    ax.set_xticks(range(len(labels)))
    ax.set_yticks(range(len(labels)))
    ax.set_xticklabels(labels)
    ax.set_yticklabels(labels)

    # Annotating each cell with the numeric value
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            # Setting colors or other conditions as needed
            color = 'black' if cm[i, j] < cm.max() / 2 else 'black'
            ax.text(j, i, format(cm[i, j], 'd'), ha='center', va='center', color=color)

    plt.show()
```
